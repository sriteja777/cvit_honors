```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
def linContrastStretching(im, a,b):

    # plt.hist(im.ravel(),256,[0,256]); plt.show()
    min_val = np.amin(im)
    max_val = np.amax(im)
    print(min_val, max_val)
    print(im.dtype)
    return ((im - min_val)*((b-a) /(max_val-min_val)) + a).astype(np.uint8)  
```

```python
def hist_equalization(im):
    # print(pr)
    out = np.zeros(im.shape)
    nop = im.shape[0] * im.shape[1]
    # print(im)
    pr = np.bincount(im.flatten(), minlength=256) / nop
    cr = np.cumsum(pr)
    # out = np.floor(255 * np.cu)
    out = np.floor(cr[im] * 255).astype(np.uint8)
    # print(pr.shape, pr[255])
    # print(out)
    # out = out.astype(np.uint8)
    # plt.figure(1)
    # plt.suptitle("Original")
    # plt.subplot(121)
    # plt.axis("off")
    # plt.imshow(im, "gray")
    # plt.subplot(122)
    # 
    # plt.hist(im.ravel(),256,[0,256])
    # 
    # plt.figure(2)
    # plt.suptitle("Histogram Equalized")
    # plt.subplot(121)
    # plt.axis("off")
    # plt.imshow(out, "gray")
    # plt.subplot(122)
    # 
    # plt.hist(out.ravel(),256,[0,256])
    
    # plt.subplot(121)
    # plt.title("PDF")
    # plt.plot(pr)
    # plt.subplot(122)
    # plt.title("CDF")
    # plt.plot(cr)
    # plt.figure(2)
    # plt.hist(im.ravel(),256,[0,256])
    # plt.figure(3)
    # plt.imshow(out, 'gray')
    # plt.show()
    return out, cr
```

```python
def plt_im_hist(im, title="Image & Histogram", fig=1):
    plt.figure(fig)
    plt.suptitle(title)
    plt.subplot(121)
    plt.axis("off")
    plt.imshow(im, "gray")
    plt.subplot(122)
    plt.ylim((0,1000))
    plt.hist(im.ravel(),256,[0,256])
    plt.show()
```

```python
def hist_matching(im, ref_img):
    # plt.figure(1)
    # plt.suptitle("Original")
    # plt.subplot(121)
    # plt.axis("off")
    # plt.imshow(im, "gray")
    # plt.subplot(122)
    # plt.hist(im.ravel(),256,[0,256])
    # 
    # plt.figure(2)
    # # plt.tight_layout()
    # plt.suptitle("Reference Image")
    # plt.subplot(121)
    # plt.axis("off")
    # plt.imshow(ref_img, "gray")
    # plt.subplot(122)

    # plt.hist(ref_img.ravel(),256,[0,256])
    out = np.copy(im)
    _, cr_im = hist_equalization(im)
    _, cr_ref_im = hist_equalization(ref_img)
    
    out_gray = np.interp(cr_im, cr_ref_im, np.arange(0, 256))
    out = out_gray[im]
    # plt_im_hist(out, title="Matched Image", fig=3)
    # print('out', out)
    return out.astype(np.uint8)
    # for  i in range(0, 256):
```

```python
def find_orig():
    parts = []
    orig = cv2.imread('canyon.png')
    orig_gray = cv2.cvtColor(orig, cv2.COLOR_BGR2GRAY)
    
    # plt_im_hist(orig_gray, title="Original", fig=1)
    parts_gray = []
    orig_parts = []
#     print("orig -> shape = ", orig_gray.shape)
    for i in range(4):
        parts.append(cv2.imread('part'+str(i+1)+'.png'))
        parts_gray.append(cv2.cvtColor(parts[-1], cv2.COLOR_BGR2GRAY))
#         print("part "+str(i)+" -> "+str(parts_gray[-1].shape))
    
    
    orig_parts.append(orig_gray[0:parts_gray[0].shape[0], 0:parts_gray[0].shape[1]])
    orig_parts.append(orig_gray[0:parts_gray[1].shape[0], parts_gray[0].shape[1]:])
    orig_parts.append(orig_gray[parts_gray[0].shape[0]:, 0:parts_gray[2].shape[1]])
    orig_parts.append(orig_gray[parts_gray[1].shape[0]:, parts_gray[2].shape[1]:])
    
#     plt_im_hist(orig_parts[0], "Orig Part 0", fig=5)
#     plt_im_hist(parts_gray[0], "Part 0", fig=6)
#     match = hist_matching(parts_gray[0], orig_parts[0])
#     plt_im_hist(match, "Matched", fig=17)
#     print(np.amin(match))
#     print(np.amax(match))
#     match, _ = hist_equalization(match)
#     plt_im_hist(match, "Stretched", fig=18)
#     print(np.amin(match))
#     print(np.amax(match))
#     return 
    matched= []
    for i in range(4):
        print(orig_parts[i].shape, parts_gray[i].shape)
        plt_im_hist(orig_parts[i], "Orig Part "+str(i), fig=6+i)
        plt_im_hist(parts_gray[i], title="part "+str(i), fig=i+2)
        matched.append(hist_matching(parts_gray[i], orig_parts[i]))
        plt_im_hist(matched[-1], title="Matched Part "+str(i), fig=10+i)
    
    final = np.concatenate(
        (np.concatenate((matched[0], matched[1]), axis=1),
        np.concatenate((matched[2], matched[3]), axis=1)),
        axis=0
    )
    plt_im_hist(orig_gray, fig=15, title="Original")
    plt_im_hist(final,fig=14, title="Final")
    
```

```python
imgs = ["car.jpeg", "ein.jpg", "wiki.jpg"]
for im in imgs:
    i = cv2.imread(im)
    i = cv2.cvtColor(i, cv2.COLOR_BGR2GRAY)
    out, _ = hist_equalization(i)
    plt_im_hist(i, title="Original")
    plt_im_hist(out, title="Histogram Equalized")
```

We can see that histograms are getting equalized and output  images have enhanced contrast. These results are better than just a linear contrast stretching. And we can see that a lot of values are missing in the histogram
