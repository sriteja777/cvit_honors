```python
def show_img(img, gray=True, splt=111, title="Image",axis="off"):
    plt.subplot(splt)
    plt.axis(axis)
    plt.title(title)
    if gray:
        plt.imshow(img, "gray",vmin=np.amin(img),vmax=np.amax(img))
    else:
        plt.imshow(img)
```
