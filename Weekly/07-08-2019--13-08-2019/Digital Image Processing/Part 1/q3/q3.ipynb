```python
import numpy as np
import cv2
import random
from matplotlib import pyplot as plt
import time
```

```python
def BitQuantizeImage(im, k):
    q_img = np.array(im, copy=True)
    num_levels = 2**k
    split_len = 256 / num_levels
    start = 0
    end = split_len
    # q_img[np.where(np.logical_and(im>=92, im<=97))] = 0
    # q_img[np.where(np.logical_and(im>97, im<105))] = 1

    for i in range(num_levels):
        q_img[np.where(np.logical_and(im>=start, im<end))] = i
#         print(q_img[0,:])
        # time.sleep(1)
        start =end
        end+= split_len
    return q_img
```

```python
def bitPlane(im, k):
    return im&(2**k)
```

```python
img = cv2.imread('quantize.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

for k in range(1, 9):
    q_img = BitQuantizeImage(img, k)
    plt.subplot(330+k)
    plt.subplots_adjust(top=2, right=2)
    plt.title(str(k) + "-bit quantized image")
    plt.axis('off')
    plt.imshow(q_img)
# plt.show()
```

```python
cm_img = cv2.imread('cameraman.png')
cm_img = cv2.cvtColor(cm_img, cv2.COLOR_BGR2GRAY)
plt.subplots_adjust(top=1.3, right=2)
for i in range(8):
    plt.subplot(241+i), plt.axis('off')
    plt.title(str(i) + "-bit plane")
    plt.imshow(bitPlane(cm_img, i), 'gray')
```
